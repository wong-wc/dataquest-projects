# Visualizing Earnings Based on College Major

In this project, the earnings based on college majors for students who graduated from college between 2010 and 2012 would be visualized by making use of visualization tools with pandas plotting functionality. The original data was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/) which is then cleaned by FiveThirtyEight. The dataset can be downloaded from FiveThirtyEight's [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors). 

Each row in the dataset corresponds to a different college major.
The data dictionary of the data set is as follows: 
- `Rank` - Rank by median earnings (the dataset is ordered by this column).
- `Major_code` - Major code.
- `Major` - Major description.
- `Major_category` - Category of major.
- `Total` - Total number of people with major.
- `Sample_size` - Sample size (unweighted) of full-time.
- `Men` - Male graduates.
- `Women` - Female graduates.
- `ShareWomen` - Women as share of total.
- `Employed` - Number employed.
- `Median` - Median salary of full-time, year-round workers.
- `Low_wage_jobs` - Number in low-wage service jobs.
- `Full_time` - Number employed 35 hours or more.
- `Part_time` - Number employed less than 35 hours.

From the visualizations, we would better be able to explore questsions relating to the data set.

```python
# Allowing visualizations to be viewed within Jupyter
%matplotlib inline

# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
# Reading in the csv file to a DataFrame: recent_grads
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
# Dropping rows with missing values
raw_data_count = recent_grads.shape[0]
recent_grads.dropna(axis='index', inplace=True)
cleaned_data_count = recent_grads.shape[0]
print('Before Cleaning: {} rows'.format(raw_data_count))
print('After Cleaning: {} rows'.format(cleaned_data_count))
```

There is one row that contains missing values and is dropped from the DataFrame `recent_grads` after cleaning.

# Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size')
```

From the plot above, there doesn't seem to be a clear connection between the sample size and the median salary. It is this inconclusive that students in more popular majors make more money.

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. Sample_size')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs. Full_time')
```

The link between the number of full-time employees and median salary is not clear from the plot above

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. ShareWomen')
```

From the plot above, there is no correlation between the percentage share of female graduates and the median salary. Hence, there appears to be no evidence of students that majored in subjects that were majority female making more money.

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Median vs. Men')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Median vs. Women')
```
